```python
texts = "NASA was established in 1958, succeeding the National Advisory Committee for Aeronautics (NACA), to give the US space development effort a distinctly civilian orientation, emphasizing peaceful applications in space science."
```

```python
for text in texts:
    sentences= nltk.sent_tokenize(texts)
    for sentence in sentences:
        words= nltk.word_tokenize(sentence)
        tagged= nltk.pos_tag(words)
        print (tagged)
```
